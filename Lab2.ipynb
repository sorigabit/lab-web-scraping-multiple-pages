```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# A list with the different kind of datasets available in data.gov.uk
```

```python
url = 'https://data.gov.uk/'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# soup
```

```python
soup.select("#content > div > div > div.uscitemlist > div:nth-child(3)")
```

```python
titles = soup.find_all(class_='govuk-heading-s dgu-topics__heading') 
for title in titles:
    print(title.get_text())
```

```python
#Display the top 10 languages by number of native speakers stored in a pandas dataframe
```

```python
url1 = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
```

```python
response1 = requests.get(url1)
response1.status_code
```

```python
soup1 = BeautifulSoup(response1.content, "html.parser")
```

```python
# soup1
```

```python
wikitable= soup1.find('table',{'class':"wikitable sortable static-row-numbers"})
# wikitable
```

```python
df=pd.read_html(str(wikitable))
df=pd.DataFrame(df[0])
print(df.head())
```

```python
df2 = df.sort_values('Native speakers (millions)', ascending=False)
df2.head(10)
```
